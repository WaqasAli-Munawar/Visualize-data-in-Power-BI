**Description of Project**

In this project we will employ the some of the Power BI skills we have learned thus far to construct a series of related reports and a dashboard from a single data source.

We will be using visualizations that include Scatter Charts and Bar Charts to create reports that will give insight about our dataset; including **Life Expectancy by Year** and **GDP Per Capita by Year**.

We will gain a firm understanding of all the phases of a full life cycle Report and Dashboard development that include Loading Data, Cleaning Data, Correcting Errors, Choosing and Configuring Vizualizations.

**Description of the dataset**

The Gapminder dataset **SampleDataWS**, combines data from multiple sources into unique coherent time-series information about life expectancy over time for countries and regions around the world.

When we loaded the data, many errors occurred. This is a normal behavior in the real-life. Data is never in perfect shape for analysis when we get it. We need to clean it. Let's start our clean and transform journey by troubleshooting the issues.

Click on **View errors** to see them.



The errors we've got are mainly due to several cells containing `NA` in the `life_exp` column. To resolve this issue, we have to change the `life_exp` data type to `Decimal Number` and then convert the error values to `null`.

There are also some errors in the `year` column. To resolve this issue, we have replace to replace errors by `1919`.

Let's continue to clean and transform our data in order to get it ready for analysis and visualization.

Here are some cleaning and transformation tasks we want to handle:

* The `Name` column actually contains country names and should be renamed to `Country`.
* The name `life_exp` is not meaningful. Rename it `Life Expectancy`.
* The `g77` column won't be used for our analysis. Let's remove it.
